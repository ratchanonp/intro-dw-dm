```python
dataset = [
    ["Hot Dogs", "Buns", "Ketchup"],
    ["Hot Dogs", "Buns"],
    ["Hot Dogs", "Coke", "Chips"],
    ["Chips", "Coke"],
    ["Chips", "Ketchup"],
    ["Hot Dogs", "Coke", "Chips"]
]
```

```python
import pandas as pd
from mlxtend.preprocessing import TransactionEncoder

te = TransactionEncoder()
te_ary = te.fit(dataset).transform(dataset)
df = pd.DataFrame(te_ary, columns=te.columns_)
df
```

```python
from mlxtend.frequent_patterns import fpgrowth

fpgrowth(df, min_support=0.33, use_colnames=True, verbose=1)
```

```python
from fpgrowth_py import fpgrowth
itemSetList = [["Hot Dogs", "Buns", "Ketchup"],
    ["Hot Dogs", "Buns"],
    ["Hot Dogs", "Coke", "Chips"],
    ["Chips", "Coke"],
    ["Chips", "Ketchup"],
    ["Hot Dogs", "Coke", "Chips"]]
freqItemSet, rules = fpgrowth(itemSetList, minSupRatio=0.33, minConf=0.6)
print(freqItemSet)

for rule in rules:
    print(rule)
```
